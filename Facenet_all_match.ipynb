```
import requests

def download_file_from_google_drive(id, destination):
    URL = "https://docs.google.com/uc?export=download"

    session = requests.Session()

    response = session.get(URL, params = { 'id' : id }, stream = True)
    token = get_confirm_token(response)

    if token:
        params = { 'id' : id, 'confirm' : token }
        response = session.get(URL, params = params, stream = True)

    save_response_content(response, destination)    

def get_confirm_token(response):
    for key, value in response.cookies.items():
        if key.startswith('download_warning'):
            return value

    return None

def save_response_content(response, destination):
    CHUNK_SIZE = 32768

    with open(destination, "wb") as f:
        for chunk in response.iter_content(CHUNK_SIZE):
            if chunk: # filter out keep-alive new chunks
                f.write(chunk)

if __name__ == "__main__":
    file_id = 'TAKE ID FROM SHAREABLE LINK'
    destination = 'DESTINATION FILE ON YOUR DISK'
    download_file_from_google_drive("18UHTJx0biyVfazaKb9hSxloy-9PNKjFo", "data.zip")
```

```
import imgaug as ia
from imgaug import augmenters as iaa
import numpy as np
import imageio

ia.seed(1)

dirs = os.listdir("cleantrain_aligned")
dirs = [x for x in dirs if "txt" not in x]
for d in dirs:
  image = os.listdir("cleantrain_aligned/"+d)
  if len(image) == 0:
    continue
  image = image[0]
  img = imageio.imread("cleantrain_aligned/"+d+"/"+image) #read you image
  images = np.array(
      [img for _ in range(7)], dtype=np.uint8)  # 32 means creat 32 enhanced images using following methods.
  sometimes = lambda aug: iaa.Sometimes(0.5, aug)
  seq = iaa.Sequential(
      [
          iaa.Fliplr(0.5),  
          iaa.Crop(percent=(0, 0.2)),            
          iaa.Sometimes(0.5, iaa.GaussianBlur(sigma=(0, 0.5))),        
          iaa.ContrastNormalization((0.75, 1.5)),         
          iaa.AdditiveGaussianNoise(
              loc=0, scale=(0.0, 0.05 * 255), per_channel=0.5),    
          iaa.Multiply((0.8, 1.2), per_channel=0.2),
          iaa.Affine(
              scale={
                  "x": (0.8, 1.2),
                  "y": (0.8, 1.2)
              },
              translate_percent={
                  "x": (-0.2, 0.2),
                  "y": (-0.2, 0.2)
              },
              shear=(-8, 8))
               
      ],
      random_order=True)  # apply augmenters in random order

  images_aug = seq.augment_images(images)

  for i in range(5):
      imageio.imwrite("cleantrain_aligned/"+d+"/"+image.split()[0]+"_"+str(i)+".png", images_aug[i])  #write all changed images
  for i in range(2):  
      if os.path.exists("cleantest_aligned/"+d):  
        imageio.imwrite("cleantest_aligned/"+d+"/"+image.split()[0]+"_"+str(i)+".png", images_aug[i])  #write all changed images
```
